### BatchNorm1d、BatchNorm2d、BatchNorm3d

For each feature, batch normalization computes the mean and variance of that feature in the mini-batch. It then subtracts the mean and divides the feature by its mini-batch standard deviation.

作者提到：

1、如果min-batch size太小的话，效果可能不好；

2、在rnn中不起作用。

另外使用BN的时候有一些陷阱。由于BN的机制，remember that batch normalization makes the loss function dependent on the batch size.尤其是在分布式训练的时候，可能会出问题。还有在做fine-tune的时候，也要注意remember that batch normalization makes the loss function dependent on the batch size. 具体可以看[原文](http://mlexplained.com/2018/11/30/an-overview-of-normalization-methods-in-deep-learning/#more-775)

```python
import torch.nn as nn
import torch

# C from an expected input of size (N, C, L) or L from input of size (N, L)
m = nn.BatchNorm1d(3)
# Input: (N,C) or (N,C,L)
# input = torch.randn(2, 3)
input = torch.FloatTensor([[2, 2, 3], [8, 5, 6]])  # N=2,C=3
# Output: (N,C) or (N,C,L) (same shape as input)
output = m(input)
print(input)
print(output)

# C from an expected input of size (N,C,H,W)
m = nn.BatchNorm2d(2)
# input=(N,C,H,W)
input = torch.randn(3, 2, 4, 4)
output = m(input)
print(input)
print(output)


# C from an expected input of size (N,C,D,H,W)
m = nn.BatchNorm3d(2)
# Input: (N,C,D,H,W)
input = torch.randn(3, 2, 2, 10, 10)
output = m(input)
print(input)
print(output)
```

### Layer Normalization
The experimental results show that layer normalization performs well for recurrent neural networks.


```python
from IPython.display import Image
Image(filename='./images/batchnorm-vs-layernorm.png')
```

```python
from IPython.display import Image
Image(filename='./images/batchnorm-vs-layernorm-2.png')
```

```python
# Input: (N, *)
input = torch.randn(20, 5, 10, 10)
# With Learnable Parameters
# normalized_shape=除了batch之外的所有维度
m = nn.LayerNorm(input.size()[1:])
# Without Learnable Parameters
m = nn.LayerNorm(input.size()[1:], elementwise_affine=False)
# Normalize over last two dimensions
m = nn.LayerNorm([10, 10])
# Normalize over last dimension of size 10
m = nn.LayerNorm(10)
# Activating the module
output = m(input)
```

### Instance Normalization
Instance normalization is similar to layer normalization but goes one step further: it computes the mean/standard deviation and normalize across each channel in each training example.
